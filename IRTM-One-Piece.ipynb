### Extract Data from One Piece fandom

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
#dataframe to store Title, volume and arc 
df = pd.DataFrame(columns=column_names)
column_names = ["Chapter", "Title", "Volume", "Arc"]
for i in range(1082):
    #Find chapter Url and read webpage data
    url = 'https://onepiece.fandom.com/wiki/Chapter_'+str(i+1) 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    #Extract the Voulme and Arc
    if i==780:
        #Volume is not metnioned in specific format in this chapter
        vol,arc = ["Volume 78","Dressrosa Arc Chapters"]
    elif i==1037:
        vol,arc = ["Volume 103","Wano Country Arc Chapters"]
    else:
        arc_ind = text.find("Arc")
        volume_ind = text.rfind("Volume", 0, arc_ind)
        vol,arc = text[volume_ind:arc_ind+12].split(", ")
    #Extract title
    viz = "Viz Title:\n"
    title_ind = text.find(viz) + len(viz)
    title_ind2 = text.find("\n",title_ind+1)
    title = text[title_ind:title_ind2]
    row = {"Chapter":i+1,"Title":title,"Volume":vol,"Arc":arc}
    df = df.append(row, ignore_index=True)
    summary_ind = text.find("Long Summary")
    summary_ind2 = text.find("Long Summary",summary_ind+1)
    summary = text[summary_ind2:]
    ref_ind = summary.find("Quick Reference")
    summary = summary[len("Long Summary "):ref_ind]
    summary = summary.replace("\n", " ")
    summary = summary.replace("\'", "")
    with open("./Chapters_summaries/"+str(i+1)+".txt", "w") as output_file:
        output_file.write(summary)
df.to_csv("chapters_info.csv", index=False)
df
```

```python
# Define the URL of the webpage
url = "https://onepiece.fandom.com/wiki/List_of_Locations"
# Send a request to the webpage and retrieve its HTML content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
locations = soup.get_text()
loc_ind = locations.find("Site Navigation")+len("Site Navigation.  ")
loc_ind2 = locations.find("God Valley")+len("God Valley")
locs= locations[loc_ind:loc_ind2].split("\n")
locs = [loc[:loc.index("(")].strip() if "(" in loc else loc for loc in locs]
locs = [loc for loc in locs if len(loc) > 0]
locs = [loc for loc in locs if "." not in loc]
#Hardcode some locations that are hard to extract
locs.append("St. Poplar")
locs.append("Mt. Colubo")
locs.append("Sandy Island")
locs.append("New Spiders Cafe")
```

### Entity Extraction: Organization
Here we attempt to extract orgainzations in One Piece but as it's shown here organization extraction in One Piece is not trivial as it follows a Hierarchichal structure.

```python
#Extracting Organizations is more complex so it requires several websites and some hardcoding
organizations = []
#Antagonist
url = "https://onepiece.fandom.com/wiki/Category:Antagonist_Groups"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("153rd Branch")
txt_ind2 = text.find("Zombie") + len("Zombie")
ant = text[txt_ind:txt_ind2].split("  ")
ant = [org for org in ant if len(org)>2 ]
ant = [org.strip() for org in ant if len(org.strip()) > 1]
organizations.extend(ant)
#Families 
url = "https://onepiece.fandom.com/wiki/Category:Families"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("Charlotte Family",text.find("Charlotte Family")+1)
txt_ind2 = text.find("Vinsmoke Family",txt_ind+1) + len("Vinsmoke Family")
fam = text[txt_ind:txt_ind2].split("  ")
fam = [org.strip() for org in fam if len(org.strip()) > 1]
fam = [org for org in fam if org[:3] != "Cat"]
organizations.extend(fam)
#Wano families
url = "https://onepiece.fandom.com/wiki/Category:Wano_Country_Families"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("Shimotsuki Family")
txt_ind2 = text.find("Hitaki Family") + len("Hitaki Family")
wano = text[txt_ind:txt_ind2].split("  ")
wano = [org.strip() for org in wano if len(org.strip()) > 1]
organizations.extend(wano)
url = "https://onepiece.fandom.com/wiki/Category:Flashback_Introduction_Groups"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("Barrels Pirates")
txt_ind2 = text.find("Tom's Workers",txt_ind+1) + len("Tom's Workers")
former = text[txt_ind:txt_ind2].split("  ")
former = [org.strip() for org in former if len(org.strip()) > 1]
organizations.extend(former)
#Former 2
url = "https://onepiece.fandom.com/wiki/Category:Former_Organizations"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("77th Branch")
txt_ind2 = text.find("Usopp Pirates",txt_ind+1) + len("Usopp Pirates")
former = text[txt_ind:txt_ind2].split("  ")
former = [org.strip() for org in former if len(org.strip()) > 1]
former =  [org for org in former if org[:3] != "Cat"]
organizations.extend(former)
#Marines
url = "https://onepiece.fandom.com/wiki/Category:Marine"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text().replace("\n", " ").replace("\t", " ")
txt_ind = text.find("Admiral")
txt_ind2 = text.find("Captain",txt_ind+1) + len("Captain")
marines = text[txt_ind:txt_ind2].split(" ")
marines[1] = marines[1] +" " +marines[2]
marines.pop(2)
organizations.extend(marines)
organizations
```
